```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datasets/scraped_final_list_URL_TripAdvisor.csv", names=["restaurant", "date", "rating", "title", "review", "url", "num_page"])
df = df.drop_duplicates(subset=["review"])
df = df[["restaurant", "date", "rating", "title", "review"]]
df
```

```python
months = ["January", "February","March","April","May","June","July","August","September","October","November","December"]
month2num = dict((k,i+1) for i, k in enumerate(months))

def convert_date_str(date):
    if date is not None:
        date = date.replace(",","")
        month_str, day, year = date.split()
        month = str(month2num[month_str])
        return "/".join([day,month,year])

convert_date_str("June 20, 2012")
```

```python
pd.to_datetime(df["date"].apply(convert_date_str))
```
