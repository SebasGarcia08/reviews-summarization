```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datasets/scraped_final_list_URL_TripAdvisor.csv", names=["restaurant", "date", "rating", "title", "review", "url", "num_page"])
df = df.drop_duplicates(subset=["review"])
df["rating"] /= 10
df
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df = df.convert_dtypes()
df.head()
```

```python
names = ["restaurant","url","inspection_date","violation_desc","score","grade","lat","long"]
df2 = pd.read_csv("../datasets/final_list_URL_TripAdvisor_noduplicated.csv")
df2 = df2.iloc[:,:-1]
df2.columns = names
df2["inspection_date"] = pd.to_datetime(df2["inspection_date"])
df2.columns = names
df2
```

```python
data = {
    "rating_before": [],
    "rating_after": []
} 

for r in df2["restaurant"].values:
    inspection_date = df2[df2["restaurant"] == r]["inspection_date"].values[0]
    restaurant_reviews = df[df["restaurant"] == r]
    rating_before = restaurant_reviews[restaurant_reviews["date"] <= inspection_date]["rating"].mean()
    rating_after = restaurant_reviews[restaurant_reviews["date"] > inspection_date]["rating"].mean()
    data["rating_before"].append(rating_before)
    data["rating_after"].append(rating_after)
```

```python
ratings_per_res = pd.DataFrame(data)
print(ratings_per_res.isna().sum())
ratings_per_res.head()
```

```python
df2[["mean_rating_before","mean_rating_after"]] = ratings_per_res.values
df2
```

```python
df2.to_csv("../datasets/mean_ratings_final_list_URL_TripAdvisor.csv", index=False)
```
